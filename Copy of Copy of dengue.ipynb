```
# read in the csv file as a dataframe


dengue_df = pd.read_csv('/content/gdrive/MyDrive/Colab Notebooks/Copy of dengue_dataset_2021-02-05(v3).csv', # just change the name if you have a new one
                        index_col='date')

dengue_df.index = pd.to_datetime(dengue_df.index, format='%Y-%m-%d')
```

```

fig, ax = plt.subplots(figsize=(15, 10))

ax.plot(dengue_df.index, dengue_df['Dengue Fever']) # change y-axis to any variable listed above


fig.autofmt_xdate()

plt.show()
```

```
fig, ax = plt.subplots(figsize=(15, 10))

ax.plot(dengue_df.index, dengue_df['Dengue Fever'], label="Dengue Fever Cases") # change y-axis to any variable listed above
ax.plot(dengue_df.index, dengue_df['Average.daily.rainfall'], label="Rainfall")
ax.plot(dengue_df.index, dengue_df['Mean.daily.tempearature'], label="Temeprature")
ax.plot(dengue_df.index, dengue_df['rh_monthly_min'], label="Relative Humidity")
ax.plot()

fig.autofmt_xdate()

plt.xlabel("Date")
plt.ylabel("Dengue Fever Cases")
plt.title("Dengue Fever Cases vs Date with Other Factors")
plt.legend()
plt.show()
```

```
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
sns.set(color_codes=True)

sns.heatmap(dengue_df, linewidths=10, annot=dengue_df, fmt="", annot_kws={"size": 12}, vmin=0, vmax=100)
```

```
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
sns.set(color_codes=True)

plt.figure(figsize=(14, 14))
plt.suptitle("Heatmap for Dengue Cases Correlation", color='g')
sns.set(font_scale=0.8)

correlation = dengue_df.corr()
sns.heatmap(correlation, annot=True)
```

```
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

sns.set(style='ticks', color_codes=True)

plt.figure(figsize=(14, 14))
plt.suptitle("Heatmap for Dengue Cases", color='b')
sns.set(font_scale=0.8)

sns.heatmap(dengue_df.astype(float).corr(), 
            linewidths=0.1, 
            square=True, 
            linecolor='white', 
            annot=True)
plt.show()
```
